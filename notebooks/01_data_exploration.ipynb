
# Data Exploration with Python

## Project Overview
This notebook demonstrates exploratory data analysis (EDA) using the surveys dataset. The analysis includes:
- Data loading and inspection
- Dataset structure assessment
- Summary statistics
- Species composition analysis
- Sex distribution analysis
- Missing value assessment
- Key observations


```

import pandas as pd
import numpy as np

surveys_df = pd.read_csv("data/surveys.csv")
surveys_df.head()
```

## Dataset Overview

```

surveys_df.info()
surveys_df.describe()
```

## Unique Sampling Sites

```

site_names = pd.unique(surveys_df["plot_id"])
site_names
```

## Summary Statistics by Sex

```

grouped_data = surveys_df.groupby('sex')
grouped_data.describe()
```

## Species Composition

```

species_counts = surveys_df.groupby('species_id')['record_id'].count()
species_counts.sort_values(ascending=False)
```

## Counts per Sampling Site

```

plot_counts = surveys_df.groupby('plot_id')['record_id'].nunique()
plot_counts.sort_values(ascending=False)
```

## Missing Values Assessment

```

for c in surveys_df.columns:
    print(c, surveys_df[c].isna().sum())
```


## Key Observations

- The dataset contains observations from multiple species and sampling sites.
- Grouping by sex allows comparison of biological measurements between males and females.
- Species counts reveal differences in representation across taxa.
- Missing values should be considered before downstream analyses.
